###  Чтение данных
Данные могут быть представлены в различных форматах:
1. CSV формат - широко распространенный формат передачи данных
2. XLSX формат - для удобной работы с эксель таблицами
3. Parquet формат - самый легкий формат передачи и хранения данных в привычном формате

Самые важные аргументы при чтении файла:

 - filepath_or_buffer (перый и единственный обязательный аргумент) --- имя файла
 - sep  разделитель (; , \t ...)
 - quotechar  символ кавычек, все что внутри считается за строку (разделители также могут входить в эту строку; ' " ...)
 - names  список названий колонок
 - header  номер строки файла (с 0), которую нужно считать заголовком
 - dtype  словарь, сопоставляющий именам колонок типы данных в них

### Apply, map, векторные функции

```python
### APPLY
def find_potential_winner(data):
    if data.RedOdds < data.BlueOdds:
        return "Red is favorite"
    elif data.RedOdds > data.BlueOdds:
        return "Blue is favorite"
    else:
        return "equal chances"
```

```python
initial_data["potential_winner"] = initial_data.apply(find_potential_winner, axis=1)
```

```python
### Векторные
initial_data.RedOdds - initial_data.BlueOdds
```

```python
### MAP
mapper = {"Red is favorite": "R", "Blue is favorite": "B", "equal chances": "E"}
```

```python
initial_data["potential_winner"].map(mapper)
```
